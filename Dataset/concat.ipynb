```python
import pandas as pd
```

```python
a=pd.read_csv('user_whiskey_interaction.csv',sep="$")
b=pd.read_csv('whiskey_images_links.csv',sep="$")
c=pd.read_csv('whiskey_w_tags.csv')
d=pd.read_csv('whiskey cluster.csv')
```

```python
temp=dict()
tmp=[]
for i in d.columns[1:]:
    min_x=d[i]-d[i].min()
    min_max=d[i].max()-d[i].min()
    tmp.append(min_x/min_max)




mins=[]
```

# Cluster Normalization

```python
def mean_norm(df_input):
    return df_input.apply(lambda x: (x-x.mean())/ x.std(), axis=0)
```

```python
temp=pd.concat([d[d.columns[0]],mean_norm(d[d.columns[1:]])],axis=1)
```

```python
temp
```

```python
clust=c.loc[0]['Cluster']
tmp_df=temp[temp['cluster']==clust][['cluster','body', 'sweet', 'sherry', 'malt', 'aperitif', 'smoky',
       'pungent', 'fruity', 'honey', 'floral', 'spicy', 'medicinal', 'nutty',
       'winey']]
for i in range(1,len(c)):
    clust=c.loc[i]['Cluster']
    tmp_df=pd.concat([tmp_df,temp[temp['cluster']==clust][['cluster','body', 'sweet', 'sherry', 'malt', 'aperitif', 'smoky',
       'pungent', 'fruity', 'honey', 'floral', 'spicy', 'medicinal', 'nutty',
       'winey']]])
```

```python
pd.concat([c[['Whiskey', 'Meta Critic', 'STDEV', '#', 'Cost', 'Class',
       'Super Cluster']],tmp_df.reset_index().drop('index',axis='columns')],axis=1).to_csv('whiskey_w_tags_norm.csv',sep="$",index=False)
```

# Links tags concat


```python
a=pd.read_csv('user_whiskey_interaction.csv',sep="$")
b=pd.read_csv('whiskey_images_links.csv',sep="$")
c=pd.read_csv('whiskey_w_tags_norm.csv',sep="$")
```

```python
whisk=c.loc[0]['Whiskey']
tmp_df=b[b['whiskey']==whisk][['images','links']]
for i in range(1,len(c)):
    whisk=c.loc[i]['Whiskey']
    tmp_df=pd.concat([tmp_df,b[b['whiskey']==whisk][['images','links']]])
```

```python
pd.concat([c,tmp_df.reset_index().drop('index',axis='columns')],axis=1).to_csv('total_df.csv',sep="$",index=False)
```
